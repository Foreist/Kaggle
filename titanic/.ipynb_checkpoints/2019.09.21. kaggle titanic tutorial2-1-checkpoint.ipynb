# 1.1 Null data check

```python
# train set check
for col in df_train.columns:
    msg = 'column: {:<10}\t Percent of NaN value: {:.2f}%'.format(col, 100 * (df_train[col].isnull().sum() / df_train[col].shape[0]))
    # 첫 번째에 컬럼명, 두 번째에 '컬럼 결측치 갯수 / 전체 row 수'
    print(msg)
```

```python
# test set check
for col in df_test.columns:
    msg = 'column: {:>10}\t Percent of NaN value: {:.2f}%'.format(col, 100 * (df_test[col].isnull().sum() / df_test[col].shape[0]))
    print(msg)
```

## 2.4 Age

```python
print('max {:.1f} Ymears'.format(df_train['Age'].max()))
print('min {:.1f} Years'.format(df_train['Age'].min()))
print('mean {:.1f} Years'.format(df_train['Age'].mean()))
```

```python
fig = plt.subplots(figsize = (9, 5))
sns.kdeplot(df_train[df_train['Survived'] == 1]['Age']) # 생존한 자들 중 나이만 보여줘
sns.kdeplot(df_train[df_train['Survived'] == 0]['Age'])
plt.legend(['Survived == 1', 'Survived == 0'])
```

```python
plt.figure(figsize = (8, 6))
df_train['Age'][df_train['Pclass'] == 1].plot(kind = 'kde')
df_train['Age'][df_train['Pclass'] == 2].plot(kind = 'kde')
df_train['Age'][df_train['Pclass'] == 3].plot(kind = 'kde')

plt.xlabel('Age')
plt.title('Age Distribution within classes')
plt.legend(['1st Class', '2nd Class', '3rd Class'])
```

### 나이에 따른 생존률을 확인하자

```python
# 해당 나이인 사람들 중 생존한 사람들의 합 / 해당 나이인 사람들
cummulate_survival_ratio = []
for i in range(1, 80):
  cummulate_survival_ratio.append(df_train[df_train['Age'] < i]['Survived'].sum() / len(df_train[df_train['Age'] < i]['Survived']))

plt.figure(figsize = (7, 7))
plt.plot(cummulate_survival_ratio)
plt.title('Survival rate change depending on range of Age', y = 1.02)
plt.ylabel('Survival rate')
plt.xlabel('Range of Age(0~x)')
```

## 2.7 Family = SibSp(형제 자매) + Parch(부모 자녀)

```python
df_train['FamilySize'] = df_train['SibSp'] + df_train['Parch'] + 1 # 자기자신 포함
df_test['FamilySize'] = df_test['SibSp'] + df_test['Parch'] + 1 # 자기자신 포함
```

```python
print('Maximum size of Family: ', df_train['FamilySize'].max())
print('Maximum size of Family: ', df_train['FamilySize'].min())
```

```python
f, ax = plt.subplots(1, 3, figsize = (40, 10))
sns.countplot('FamilySize' , data = df_train, ax = ax[0])
ax[0].set_title('(1) No. Of Passengers Boarded', y = 1.02)

sns.countplot('FamilySize', hue = 'Survived', data = df_train, ax = ax[1])
ax[1].set_title('(2) Survived countplot depending on FamilySize', y = 1.02)

# 인원수에 따른 생존확률
df_train[['FamilySize', 'Survived']].groupby(['FamilySize'], as_index = True).mean().sort_values(by = 'Survived', ascending = False).plot.bar(ax = ax[2])
ax[2].set_title('(3) Survived rate depending on FamilySize', y = 1.02)

plt.subplots_adjust(wspace = 0.2, hspace = 0.5)
plt.show()
```

## 2.8 Fare

```python
fig, ax = plt.subplots(1, 1, figsize = (8, 8))
sns.distplot(df_train['Fare'], color = 'b', label = 'Skewness: {:.2f}'.format(df_train['Fare'].skew()), ax = ax)
plt.legend(loc = 'best')
# 이건 너무 비대칭적이라서 log를 취해서 대칭적으로 만들어줘야함
```

```python
df_test.loc[df_test.Fare.isnull(), 'Fare'] = df_test['Fare'].mean()
df_train['Fare'] = df_train['Fare'].map(lambda i: np.log(i) if i > 0 else 0)
df_test['Fare'] = df_test['Fare'].map(lambda i: np.log(i) if i > 0 else 0)
```

```python
fig, ax = plt.subplots(1, 1, figsize = (8, 8))
g = sns.distplot(df_train['Fare'], color = 'b', label = 'Skewness: {:.2f}'.format(df_train['Fare'].skew()), ax = ax)
g = g.legend(loc = 'best')
```
